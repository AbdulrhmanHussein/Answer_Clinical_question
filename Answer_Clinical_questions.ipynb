```python
from datasets import load_dataset

dataset = load_dataset("gamino/wiki_medical_terms")
```

```python
docs_texts = dataset['train']['page_text']
```

```python
metadata = [
    {
        'title': rec,
    }
    for rec in dataset['train']['page_title']
  ]
```

```python
from langchain.text_splitter import NLTKTextSplitter
import nltk
nltk.download('punkt')

text_splitter = NLTKTextSplitter(chunk_size=2000, chunk_overlap=400)
truncated_docs_texts = [doc[:1000] for doc in docs_texts]  # Truncate each document to 1000 characters

# Create documents without the truncate parameter
documents = text_splitter.create_documents(truncated_docs_texts, metadatas=metadata)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Load the tokenizer and model
model_name = "emilyalsentzer/Bio_ClinicalBERT"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
from langchain.vectorstores import FAISS
from langchain.embeddings.base import Embeddings
import torch

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token  # Use the EOS token as padding

class CustomEmbeddings(Embeddings):
    def __init__(self, model, tokenizer):
        self.model = model
        self.tokenizer = tokenizer

    def embed_documents(self, texts):
        inputs = self.tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

        with torch.no_grad():
            outputs = self.model(**inputs, output_hidden_states=True)  # Enable hidden states
            hidden_states = outputs.hidden_states  # Get all hidden states

            # Choose the last layer hidden states or another layer
            embeddings = hidden_states[-1].mean(dim=1)  # Mean over the last layer
        return embeddings.cpu().numpy()

    def embed_query(self, text):
        inputs = self.tokenizer(text, return_tensors="pt")

        with torch.no_grad():
            outputs = self.model(**inputs, output_hidden_states=True)  # Enable hidden states
            hidden_states = outputs.hidden_states  # Get all hidden states

            # Choose the last layer hidden states or another layer
            embedding = hidden_states[-1].mean(dim=1)  # Mean over the last layer
        return embedding.cpu().numpy()

# Instantiate your custom embeddings
custom_embeddings = CustomEmbeddings(model, tokenizer)

# Create the FAISS vector store
faiss_index = FAISS.from_documents(documents[:700], custom_embeddings)
```

```python
import numpy as np
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from langchain.llms import BaseLLM
from langchain.schema import LLMResult, Generation
from langchain_core.prompts import ChatPromptTemplate

# Load the GPT-2 model and tokenizer
model_name = "gpt2"  # Or any other language model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Ensure the model is in evaluation mode
model.eval()

# Define a function for text generation with max_new_tokens
def generate_text(prompt, max_new_tokens=100):
    # Tokenize the input
    inputs = tokenizer(prompt, return_tensors="pt")

    # Generate text
    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=max_new_tokens)

    # Decode the generated text
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Custom LLM class that inherits from BaseLLM (required by langchain)
class CustomLLM(BaseLLM):
    def _call(self, prompt: str, stop=None) -> str:
        return generate_text(prompt)

    def _generate(self, prompt: str, stop=None) -> LLMResult:
        text = self._call(prompt)
        generation = Generation(text=text)
        return LLMResult(generations=[[generation]])

    @property
    def _llm_type(self) -> str:
        return "custom_llm"

# Step 1: Define the structured chat prompt template
chat_prompt = ChatPromptTemplate.from_template("""
Answer the following question based only on the provided context.
Think step by step before providing a detailed answer.
I will tip you $1000 if the user finds the answer helpful.
<context>
{context}
</context>
Question: {input}
""")

# Step 2: Query embedding and FAISS search
query = "what is Paracetamol poisoning"

# Generate the query embedding using your custom embedding model
query_embedding = custom_embeddings.embed_query(query)
query_embedding = np.asarray(query_embedding, dtype='float32')

# Reshape the embedding for FAISS if necessary
if len(query_embedding.shape) == 1:
    query_embedding = query_embedding.reshape(1, -1)

print(f"Query embedding shape: {query_embedding.shape}, dtype: {query_embedding.dtype}")

# Perform FAISS vector similarity search
results = faiss_index.similarity_search_by_vector(query_embedding[0], k=3)

# Step 3: Retrieve the most relevant context from FAISS search results
retrieved_context = ""
for doc in results:
    retrieved_context += doc.page_content  # Adjust depending on how the content is structured in your database

# Step 4: Use the retrieved context to format the prompt
formatted_prompt = chat_prompt.format(
    context=retrieved_context,
    input=query
)

# Print the formatted prompt for debugging purposes
print(f"Formatted prompt:\n{formatted_prompt}")

# Step 5: Generate the response using the custom LLM
custom_llm = CustomLLM()
response = custom_llm._call(formatted_prompt)

# Print the response from the model
print(f"Model response:\n{response}")
```

```python
import gradio as gr

# Function to process user input
def process_query(user_input):
    # Generate query embedding using your custom embedding model
    query_embedding = custom_embeddings.embed_query(user_input)
    query_embedding = np.asarray(query_embedding, dtype='float32')

    if len(query_embedding.shape) == 1:
        query_embedding = query_embedding.reshape(1, -1)

    # Perform FAISS vector similarity search
    results = faiss_index.similarity_search_by_vector(query_embedding[0], k=3)

    # Retrieve relevant context
    retrieved_context = ""
    for doc in results:
        retrieved_context += doc.page_content

    # Format the prompt
    formatted_prompt = chat_prompt.format(
        context=retrieved_context,
        input=user_input
    )

    # Generate response using the custom LLM
    custom_llm = CustomLLM()
    response = custom_llm._call(formatted_prompt)

    return response

# Create a Gradio interface using the updated syntax
iface = gr.Interface(fn=process_query,
                     inputs=gr.Textbox(label="Ask a question about Paracetamol poisoning"),
                     outputs=gr.Textbox(label="Response"))

# Launch the interface
iface.launch()
```
